**PROBLEM STATEMENT**: Image to Pencil Sketch with Python

We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image.

```python
import cv2 
import matplotlib.pyplot as plt
from PIL import Image
import os
```

Orginal Image

```python
img1 = cv2.imread('photo.jpg')
cv2.imshow('OG Image' , img1);
cv2.waitKey(0)                 #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

Grey Scale Image

```python
gray_img = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray-Scale Image' ,gray_img);
cv2.waitKey(0)                                         #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

Inverted Grey Scale

```python
invert_img = cv2.bitwise_not(gray_img)
cv2.imshow('Inverted Gray-scale Image',invert_img)
cv2.waitKey(0)                 #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

Blur Image

```python
# Blur the image
smooth_img = cv2.GaussianBlur(invert_img,(23,23),sigmaX=0, sigmaY=0)
cv2.imshow('Inverted Blurry Image' , smooth_img );
cv2.waitKey(0)                 #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

Final Pencil Image

```python
final_img = cv2.divide(gray_img, 255-smooth_img, scale=250)
cv2.imshow('Final Sketch' , final_img );
cv2.waitKey(0)                 #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(7,7))
img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.title('original image')

plt.figure(figsize=(7,7))
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.title('Grayscale image')

plt.figure(figsize=(7,7))
plt.imshow(invert_img,cmap='gray')
plt.axis("off")
plt.title('Negative image')

plt.figure(figsize=(7,7))
plt.imshow(smooth_img, cmap='gray')
plt.axis("off")
plt.title('Blurred image')

plt.figure(figsize=(7,7))
plt.imshow(final_img, cmap='gray')
plt.axis("off")
plt.title('Pencil Sketch image')
plt.show()
```
